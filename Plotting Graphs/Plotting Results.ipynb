```python
import json
import os
def read_and_aggregate_json_files(directory, test_name):
    results = []
    # List all JSON files in the specified directory
    for filename in os.listdir(directory):
        if filename.endswith('.json'):
            filepath = os.path.join(directory, filename)
            with open(filepath, 'r') as file:
                data = json.load(file)
                data['model_name'] = filename.replace('.json', '')  # Add a model name key based on the file name
                results.append(data)
    save_aggregated_results(results, test_name)
    return results

def save_aggregated_results(results, output_filename):
    with open(output_filename, 'w') as file:
        json.dump(results, file, indent=4)

# Usage
directory = './results/CNN/general/'  # Change this to the path where your JSON files are stored
aggregated_results = read_and_aggregate_json_files(directory, "LSTM_general.json")

directory = './results/CNN/MGT/'  # Change this to the path where your JSON files are stored
aggregated_results = read_and_aggregate_json_files(directory, "LSTM_MGTBench.json")

directory = './results/CNN/Mix/'  # Change this to the path where your JSON files are stored
aggregated_results = read_and_aggregate_json_files(directory, "LSTM_Mix.json")
```

```python
import matplotlib.pyplot as plt
import numpy as np
from math import pi


import matplotlib.pyplot as plt
import numpy as np

def make_radar_chart(title, labels, data, models):
    # Number of variables we're plotting.
    num_vars = len(labels)

    # Compute angle each bar is centered on:
    angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
    angles += angles[:1]  # Complete the loop

    fig, ax = plt.subplots(figsize=(10, 10), subplot_kw=dict(polar=True))
    
    # Draw one line per model
    for dat, model in zip(data, models):
        values = dat + dat[:1]
        ax.plot(angles, values, 'o-', label=model)  # 'o-' adds circular markers to each point
        ax.fill(angles, values, alpha=0.25)  # Fill the area under the curve
    
    # Labels for each point
    ax.set_xticks(angles[:-1])
    ax.set_xticklabels(labels, fontsize=14, color='darkblue')

    # Use custom adjustments for label alignment to avoid overlap
    for label, angle in zip(ax.get_xticklabels(), angles):
        if angle in (0, np.pi):
            label.set_horizontalalignment('center')
        elif 0 < angle < np.pi:
            label.set_horizontalalignment('left')
        else:
            label.set_horizontalalignment('right')

    # Increase the plot's radial distance to fit the labels
    ax.set_ylim(0, 1.1)

    # Title and legend adjustments for better fit and readability
    plt.title(title, size=20, color='black', y=1.1)
    plt.legend(loc='upper right', bbox_to_anchor=(0.1, 0.1), fontsize=12)

    plt.show()

```

```python
import json
import pandas as pd
import matplotlib.pyplot as plt

def loading_data(file_name):
    with open(file_name, 'r') as file_name:
        models_data = json.load(file_name)
    # Create data frames
    overall_accuracies = []
    domain_accuracies = []
    model_accuracies = []
    detail_scores = []
    
    for model in models_data:
        # Loading the basic information
        detail_scores.append({
            "Model": model["model_name"],
            "Accuracy": model["overall_accuracy"],
            "Precision": model["precision"],
            "Recall": model["recall"],
            "F1": model["f1_score"],
            "AUC": model["auc"]
        })
        
        # Overall accuracy for each model
        overall_accuracies.append({
            "Model": model["model_name"],
            "Overall Accuracy": model["overall_accuracy"]
        })
    
        # Domain accuracies for each model
        for domain, accuracy in model["domain_accuracies"].items():
            domain_accuracies.append({
                "Model": model["model_name"],
                "Domain": domain,
                "Accuracy": accuracy
            })
    
        # Sub-model accuracies for each model
        for sub_model, accuracy in model["model_accuracies"].items():
            model_accuracies.append({
                "Model": model["model_name"],
                "Sub-Model": sub_model,
                "Accuracy": accuracy
            })
    return overall_accuracies, domain_accuracies, model_accuracies, detail_scores

def plot_overall_accuracy(df_overall):
    plt.figure(figsize=(10, 6))
    
    # Generating a color map that has as many colors as there are models
    colors = plt.cm.viridis(np.linspace(0, 1, len(df_overall["Model"])))

    # Plotting each bar with a distinct color
    for i, (model, accuracy) in enumerate(zip(df_overall["Model"], df_overall["Overall Accuracy"])):
        plt.bar(model, accuracy, color=colors[i])

    plt.xlabel('Models', fontsize=14)
    plt.ylabel('Overall Accuracy', fontsize=14)
    plt.title('Comparison of Overall Model Accuracy', fontsize=16)
    
    min_accuracy = df_overall["Overall Accuracy"].min()
    max_accuracy = df_overall["Overall Accuracy"].max()
    margin = (max_accuracy - min_accuracy) * 0.1  # Dynamic margin to enhance visibility
    
    plt.ylim([max(min_accuracy - margin, 0), min(max_accuracy + margin, 1)])  # Dynamic limits
    plt.xticks(rotation=45, fontsize=12)
    plt.tight_layout()
    plt.show()


def plot_domain_accuracy(df_domains):
    plt.figure(figsize=(14, 7))
    domains = df_domains["Domain"].unique()
    models = df_domains["Model"].unique()
    
    n_bars = len(models)
    bar_width = 0.8 / n_bars
    
    index = np.arange(len(domains))
    colors = plt.cm.viridis(np.linspace(0, 1, len(models)))
    
    for i, model in enumerate(models):
        accuracies = [df_domains[(df_domains["Domain"] == domain) & (df_domains["Model"] == model)]["Accuracy"].values[0]
                      if not df_domains[(df_domains["Domain"] == domain) & (df_domains["Model"] == model)].empty else 0 for domain in domains]
        plt.bar(index + i * bar_width, accuracies, bar_width, label=model, color=colors[i])
    
    plt.xlabel('Domains', fontsize=14)
    plt.ylabel('Accuracy', fontsize=14)
    plt.title('Domain Accuracy by Model',  fontsize=16)
    plt.xticks(index + bar_width * (n_bars - 1) / 2, domains, fontsize=14)
    #plt.legend(title="Models")
    plt.tight_layout()
    plt.show()
    
    # Plotting Latex 
    data = []
    model_names = []

    for model in models:
        model_names.append(model)
        model_data = []
        for domain in domains:
            subset = df_domains[(df_domains["Model"] == model) & (df_domains["Domain"] == domain)]
            accuracy = subset["Accuracy"].values[0] if not subset.empty else 0
            model_data.append(f"{accuracy:.2f}")  # Format for better readability in LaTeX
        data.append(model_data)

    # Normalize data by the maximum value in the dataset for better visualization
    max_value = max(max(float(d) for d in model_data) for model_data in data)
    normalized_data = [[float(x) / max_value for x in d] for d in data]

    # Create DataFrame for tabular results
    df_tabular = pd.DataFrame(data, columns=domains, index=model_names)
    df_tabular.index.name = 'Model'
    print("Tabular Results for Domain Accuracy (LaTeX):\n", df_tabular.to_latex(escape=False))

def plot_model_accuracy(df_models):
    plt.figure(figsize=(14, 7))
    sub_models = df_models["Sub-Model"].unique()
    models = df_models["Model"].unique()
    
    n_bars = len(models)
    bar_width = 0.8 / n_bars
    
    index = np.arange(len(sub_models))
    colors = plt.cm.viridis(np.linspace(0, 1, len(models)))
    
    for i, model in enumerate(models):
        accuracies = [df_models[(df_models["Sub-Model"] == sub_model) & (df_models["Model"] == model)]["Accuracy"].values[0]
                      if not df_models[(df_models["Sub-Model"] == sub_model) & (df_models["Model"] == model)].empty else 0 for sub_model in sub_models]
        plt.bar(index + i * bar_width, accuracies, bar_width, label=model, color=colors[i])
    
    plt.xlabel('Sub-Models', fontsize=14)
    plt.ylabel('Accuracy', fontsize=14)
    plt.title('Sub-Model Accuracy by Base Model', fontsize=16)
    plt.xticks(index + bar_width * (n_bars - 1) / 2, sub_models, fontsize=14)
    #plt.legend(title="Models")
    plt.tight_layout()
    plt.show()
    
    models = df_models["Model"].unique()
    sub_models = df_models["Sub-Model"].unique()
    
    data = []
    model_names = []
    
    # Normalize data by the maximum accuracy to fit radar chart scale of 0-1
    max_accuracy = df_models["Accuracy"].max()

    for model in models:
        accuracies = []
        for sub_model in sub_models:
            subset = df_models[(df_models["Model"] == model) & (df_models["Sub-Model"] == sub_model)]
            accuracy = subset["Accuracy"].values[0] / max_accuracy if not subset.empty else 0
            accuracies.append(f"{accuracy:.2f}")  # Format for better readability in LaTeX
        data.append(accuracies)
        model_names.append(model)  # Keep track of model names for legend

    labels = sub_models.tolist()

    # Create DataFrame for tabular results
    df_tabular = pd.DataFrame(data, columns=sub_models, index=model_names)
    df_tabular.index.name = 'Model'
    print("Tabular Results for Sub-Model Accuracy (LaTeX):\n", df_tabular.to_latex(escape=False))


def display_performance_table(detail_scores):
    # Convert detail_scores into a DataFrame
    df_detail_scores = pd.DataFrame(detail_scores)
    
    # Print the DataFrame in a readable format
    print("Detailed Performance Metrics:")
    print(df_detail_scores)
    
    # Convert and print the DataFrame to LaTeX format
    print("\nLaTeX Table Output:")
    print(df_detail_scores.to_latex(index=False, float_format="%.2f", escape=False))

items = ["/kaggle/input/aist4010-project-plottingdata/SVM_general.json", "/kaggle/input/aist4010-project-plottingdata/SVM_MGTBench.json", "/kaggle/input/aist4010-project-plottingdata/SVM_Mix.json"]

for item in items:
    print(f"Dealing with file {item}")
    overall_accuracies, domain_accuracies, model_accuracies, detail_accuracies = loading_data(item)
    df_overall = pd.DataFrame(overall_accuracies)
    df_domains = pd.DataFrame(domain_accuracies)
    df_models = pd.DataFrame(model_accuracies)
    plot_overall_accuracy(df_overall)
    plot_domain_accuracy(df_domains)
    plot_model_accuracy(df_models)
    display_performance_table(detail_accuracies)
```

## MixSet Plotting

```python
def loading_data(file_name):
    with open(file_name, 'r') as file_name:
        models_data = json.load(file_name)
    # Create data frames
    dataset_accuracies = []
    
    for model in models_data:    
        # Dataset accuracies for each model
        for dataset, accuracy in model["dataset_accuracies"].items():
            dataset_accuracies.append({
                "Model": model["model_name"],
                "Dataset": dataset,
                "Accuracy": accuracy
            })
    return dataset_accuracies


def plot_dataset_accuracy(df_datasets):
    plt.figure(figsize=(14, 7))
    datasets = df_datasets["Dataset"].unique()
    models = df_datasets["Model"].unique()
    
    # Determine the number of bars per group
    n_bars = len(models)
    bar_width = 0.8 / n_bars  # The width of a single bar
    
    # Create an index for the dataset groups and a color palette for the models
    index = np.arange(len(datasets))
    colors = plt.cm.viridis(np.linspace(0, 1, len(models)))
    
    # Adjusted plotting each model's dataset accuracy
    adjusted_accuracies = []
    for i, model in enumerate(models):
        accuracies = []
        for dataset in datasets:
            subset = df_datasets[(df_datasets["Dataset"] == dataset) & (df_datasets["Model"] == model)]
            if not subset.empty:
                accuracy = subset["Accuracy"].values[0]
                # Adjust accuracy for specific datasets
                
                # *************#
                # Notice that due to the classification goal in MixSet, the original accuracy recorded in those 3 mixing method is wrong
                # The correct one is 1 - oldAcc
                if dataset in ['Rewrite', 'Complete', 'Polish']:
                    accuracy = 1 - accuracy
                accuracies.append(accuracy)
            else:
                accuracies.append(0)  # Default to 0 if no data available
        adjusted_accuracies.append(accuracies)
        plt.bar(index + i * bar_width, accuracies, bar_width, label=model, color=colors[i])
    
    plt.xlabel('Datasets', fontsize=14)
    plt.ylabel('Accuracy', fontsize=14)
    plt.title('Adjusted Accuracy Over Different Methods by Model', fontsize=16)
    plt.xticks(index + bar_width * (n_bars - 1) / 2, datasets, fontsize=12)  # Center x-tick labels between the grouped bars
    plt.legend(title="Models", fontsize=12)
    plt.tight_layout()
    plt.show()
    
    # Data for radar chart and tabular results
    data = adjusted_accuracies
    model_names = models
    
    # Normalize data by the maximum value in the data set for better visualization
    max_value = max(max(d) for d in data)
    normalized_data = [[x / max_value for x in d] for d in data]
    
    make_radar_chart("Model Performance on Different Methods", datasets, normalized_data, model_names)
    
    # Create DataFrame for tabular results
    df_tabular = pd.DataFrame(data, columns=datasets, index=model_names)
    df_tabular.index.name = 'Model'
    
    # Print the DataFrame in a readable format
    print("Tabular Results for Method Accuracy:")
    print(df_tabular)
    
    # Convert and print the DataFrame to LaTeX format
    print("\nLaTeX Table Output:")
    print(df_tabular.to_latex(index=True, float_format="%.2f", escape=False))
    

items = ["/kaggle/input/aist4010-project-plottingdata/SVM_Mix.json"]

for item in items:
    print(f"Dealing with file {item}")
    dataset_accuracies = loading_data(item)
    df_datasets = pd.DataFrame(dataset_accuracies)
    plot_dataset_accuracy(df_datasets)
```
